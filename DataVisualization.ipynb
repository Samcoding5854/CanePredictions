```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python

# Read the data into a DataFrame
data = pd.read_csv('sugarcane_prices_data.csv')
```

```python
import pandas as pd
import seaborn as sns

# One-hot encode the 'Season' column
data_encoded = pd.get_dummies(data, columns=['Season'])

# Selecting columns for correlation matrix (excluding 'Year' and 'Price')
correlation_data = data_encoded.drop(['Year', 'Price'], axis=1)

# Calculating correlation matrix
correlation_matrix = correlation_data.corr()

# Plotting correlation matrix as heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Parameters (Excluding Year and Price)')
plt.show()
```

```python
# Parameter Visualization
# Plotting minimum and maximum temperature over the years for each season
seasons = data['Season'].unique()
plt.figure(figsize=(12, 6))
for season in seasons:
    season_data = data[data['Season'] == season]
    plt.plot(season_data['Year'], season_data['Min_Rainfall'], label=f'{season} Min Rainfall', marker='o')
    plt.plot(season_data['Year'], season_data['Max_Rainfall'], label=f'{season} Max Rainfall', marker='o')
plt.title('Minimum and Maximum Rainfall Over the Years for Each Season')
plt.xlabel('Year')
plt.ylabel('Rainfall (mm)')
plt.legend()
plt.grid(True)
plt.xticks(data['Year'].unique())  # Setting x-axis ticks to only show years
plt.show()
```

```python
# Parameter Visualization
# Plotting minimum and maximum temperature over the years for each season
seasons = data['Season'].unique()
plt.figure(figsize=(12, 6))
for season in seasons:
    season_data = data[data['Season'] == season]
    plt.plot(season_data['Year'], season_data['Min_Temperature'], label=f'{season} Min Temp', marker='o')
    plt.plot(season_data['Year'], season_data['Max_Temperature'], label=f'{season} Max Temp', marker='o')
plt.title('Minimum and Maximum Temperature Over the Years for Each Season')
plt.xlabel('Year')
plt.ylabel('Temperature (°C)')
plt.legend()
plt.grid(True)
plt.xticks(data['Year'].unique())  # Setting x-axis ticks to only show years
plt.show()

```

```python
data = pd.read_csv('Lucknow_data.csv')
# Grouping data by season
season_groups = data.groupby('Season')

# Plotting price fluctuation for each season
plt.figure(figsize=(10, 6))
for season, season_data in season_groups:
    plt.plot(season_data['Year'], season_data['Price'], label=season)

plt.title('Price Fluctuation in Lucknow Over the Years for Each Season')
plt.xlabel('Year')
plt.ylabel('Price')
plt.legend()
plt.grid(True)
plt.xticks(data['Year'].unique())  # Setting x-axis ticks to only show years
plt.show()
```
