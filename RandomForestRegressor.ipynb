```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt
```

```python
# Load the dataset
data = pd.read_csv('synthesized_data.csv')
```

```python
# Extract input and output columns
X = data[['Min_Temperature', 'Max_Temperature', 'Min_Rainfall', 'Max_Rainfall', 'Temperature', 'Rainfall']]
y = data['Price']
```

```python
# Normalize the input features
scaler_X = MinMaxScaler()
X_scaled = scaler_X.fit_transform(X)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

```python
# Define the Random Forest Regressor model
rf_regressor = RandomForestRegressor(n_estimators=100, random_state=42)
```

```python
# Train the model
rf_regressor.fit(X_train, y_train)
```

```python
# Make predictions
y_pred = rf_regressor.predict(X_test)
```

```python
# Calculate Mean Squared Error
mse = mean_squared_error(y_test, y_pred)
print("Mean Squared Error:", mse)
```

```python
# Plot actual vs predicted prices with custom colors
plt.figure(figsize=(12, 6))
plt.plot(y_test.values, label='Actual', color='green')  # Actual prices in blue
plt.plot(y_pred, label='Predicted', color='red')  # Predicted prices in red
plt.xlabel('Sample')
plt.ylabel('Price')
plt.legend()
plt.title('Random Forest Regressor')
plt.show()
```
